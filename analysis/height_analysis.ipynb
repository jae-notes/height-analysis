```python
import pandas as pd

df = pd.read_csv("../data/people.csv")
df
```

## Load the dataset
We load the dataset from the 'data' folder and display it to understand its structure.

```python
overall_avg = df["height"].mean()
overall_avg
```

## Overall average height
This value represents the average height of all people in the dataset.
It serves as a baseline for later comparisons.

```python
adults_avg = df[df["age"] >= 18]["height"].mean()
adults_avg
```

## Average height of adults
This value represents the average height of people aged 18 and older.
We compare this to the overall average height to understand differences between groups.

```python
adults_avg > overall_avg
```

This confirms that the average height of adults is greater than the overall average height.

```python
import matplotlib.pyplot as plt

plt.hist(df["height"], bins=5)

plt.axvline(overall_avg, linestyle="--", label="Overall average")
plt.axvline(adults_avg, linestyle=":", label="Adults average")

plt.xlabel("Height (cm)")
plt.ylabel("Number of people")
plt.title("Height distribution")

plt.legend()
plt.show()
```

### Height distribution

The histogram shows the distribution of heights in the dataset.
The vertical lines indicate the overall average height and the average height of adults (age â‰¥ 18).

The adult average height is higher than the overall average, indicating that including children lowers the overall mean height.
